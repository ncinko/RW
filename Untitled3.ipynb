```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
num_links = 100
chain_links = np.zeros((num_links,2))

for i in range(num_links):
    chain_links[i,:] = [0.1*i, 0]
```

```python
plt.scatter(chain_links[:, 0], chain_links[:,1])
plt.show()
```

```python
chain_v = np.zeros((num_links,2))
T = 100
G = 0.1
dt = 0.01
a = np.zeros((num_links,2))
for j in range(1000):
    for link in range(num_links):
        if link == 0 or link == 99:
            a[link,:] = [0,0]
        else:
            theta_l = np.arctan((chain_links[link - 1, 1] - chain_links[link , 1])/(-chain_links[link - 1, 0] + chain_links[link , 0]))
            theta_r = np.arctan((chain_links[link + 1, 1] - chain_links[link , 1])/(chain_links[link + 1, 0] - chain_links[link , 0]))
            a[link,0] = -T*np.cos(theta_l) + T*np.cos(theta_r)
            a[link,1] = -G + T*np.sin(theta_l) + T*np.sin(theta_r)
    chain_v += dt*a
    chain_links += chain_v*dt
    
```

```python
plt.scatter(chain_links[:, 0], chain_links[:,1])
a = 53
xs = np.arange(0,10,0.1)
plt.plot(xs, (a*np.cosh((xs - 5)/a) - a*np.cosh(5/a)), 'r-')
plt.show()
```
