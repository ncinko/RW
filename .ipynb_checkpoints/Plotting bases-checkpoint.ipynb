```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''Load a dictionary of 300 basis functions, each with
dimensions of 16 x 16 pixels. (1.6 times overcomplete?)'''

bases = np.load('basis1.npz')['basis']
```

```python
num_images = 15
pixels = 16

fig, axes = plt.subplots(nrows=num_images, ncols=num_images, sharex=True, sharey=True, figsize = (20,20), facecolor = 'white')

axes = axes.ravel()
for i in range(num_images**2):      
    axes[i].imshow(np.reshape(bases[:,i], (pixels, pixels)), cmap = 'plasma')
    axes[i].axis('off')
    
plt.tight_layout(h_pad = 1, w_pad = 1)
plt.show()
```
